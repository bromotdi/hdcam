```python
# the function takes a fasta file and extracts the genome as a string
def extractFasta(path_to_file):
    sequence = ''
    with open(path_to_file, mode='r') as handle:
        for record in SeqIO.parse(handle, 'fasta'):
            identifier = record.id
            description = record.description
            sequence = record.seq
            amount_of_nucleotides = len(sequence)

    sequence = str(sequence)
    return sequence


# the function cuts the sequence to kmers of length k and step
# returns a list/set with all the sequences (can convert to set by set(list1))
def kmers(k, step, sequence):
    # string_name[start:end:step]
    list1 = []
    for index in range(0, len(sequence), step):
        # print(sequence[index:index+k:step],'\n')
        # this if is for removing the extra kmers that are created when reaching the end of the sequence
        # when the length is smaller than k it means we reached the end, and the small sequences created
        # are already covered in the last kmer
        if len(sequence[index:index + k:1])<k:
            pass
        else:
            list1.append(sequence[index:index + k:1])
    return list1
```

```python
import random

def replace_N_with_random(sequence):
    result = []
    nucleotides = ["A", "G", "C", "T"]

    for char in sequence:
        if char == "N":
            result.append(random.choice(nucleotides))
        else:
            result.append(char)

    return ''.join(result)

# # Example usage:
# input_sequence = "ATNGCATN"
# new_sequence = replace_N_with_random(input_sequence)
# print(new_sequence)
```

```python
# Encode genome to gray code <------------------------NOT GRAY CODE, DIFFERENT ENCODING INSTEAD

def encode_to_gray(strings):
    gray_mapping = {'A': '000', 'G': '011', 'C': '101', 'T': '110'}
    encoded_list = []
    for string in strings:
        gray_code = ""
        for char in string:
            gray_code += gray_mapping[char]
        encoded_list.append(gray_code)
    return encoded_list

                    
encoded_list = encode_to_gray(wuhan_set_of_kmers)      # wuhan       
                
encoded_list2 = encode_to_gray(LC581376_set_of_kmers)  # other sars cov2             
                       
encoded_list3 = encode_to_gray(rota_set_of_kmers)      # different pathogen         
```
